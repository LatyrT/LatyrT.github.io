# 3. All files in 1

# 3.1. Combine all files in one

```python
Year_2019_df = pd.concat(
    [ 
        Jan_DF,
        Feb_DF,
        Mar_DF,
        Apr_DF,
        May_DF,
        Jun_DF,
        Jul_DF,
        Aug_DF,
        Sep_DF,
        Oct_DF,
        Nov_DF,
        Dec_DF
    ]
)
Year_2019_df.head()
```

# 3.2. Verify data & Change data type

```python
len(Jan_DF) + len(Feb_DF) + len(Mar_DF) + len(Apr_DF) + len(May_DF) + len(Jun_DF) + len(Jul_DF) + len(Aug_DF) + len(Sep_DF) + len(Oct_DF) + len(Nov_DF) + len(Dec_DF)
```

```python
len(Year_2019_df)
```

```python
Year_2019_df.columns
```

```python
Year_2019_df['usertype'] = Year_2019_df['usertype'].astype(str)
```

```python
Year_2019_df['birth year'] = Year_2019_df['birth year'].astype(str)
```

```python
Year_2019_df['gender'] = Year_2019_df['gender'].astype(str)
```

```python
Year_2019_df['index'] = Year_2019_df.index.astype(str)
```

```python
Year_2019_df['User_Type_ID'] = Year_2019_df['usertype'] + "-" + Year_2019_df['birth year'] + "-" +  Year_2019_df['gender']
```

# 3.3. Station 1 data

```python
Station_1_df = Year_2019_df[['start station id',
                         'start station name', 
                         'start station latitude',
                         'start station longitude']]
Station_1_df
```

```python
Station_1_df = Station_1_df.drop_duplicates(subset='start station id', keep="first")
Station_1_df
```

```python
# Export to CSV
Station_1_df.to_csv("./04_Data_Source_Tableau/Station_1.csv",
                   index=False, 
                   header=True
                  )
```
